#### ¿Cómo varía el valor de los pedidos, desde los más económicos hasta los más costosos?

```python
# Distribución del valor de los pedidos
order_value = items.groupby('order_id')['price'].sum().reset_index().rename(columns= {'price':'total_price_by_order'})

print(F'Distribución del valor de los pedidos:')
order_value['total_price_by_order'].describe()
```

- Count (conteo): Hay 98,666 pedidos en total (contabilizadados en la tabla 'items').

- Mean (promedio): El precio promedio de los pedidos es de aproximadamente R$137.75.

- Std (desviación estándar): La dispersión de los precios alrededor del promedio es de aproximadamente R$210.65. Esto indica la variabilidad de los precios de los pedidos.

- Min (mínimo): El precio mínimo de un pedido es de R$ 0.85. Esto podría representar pedidos con precios muy bajos o incluso pedidos gratuitos.

- 25% (percentil 25): El 25% de los pedidos tienen un precio total de hasta R$45.90.

- 50% (percentil 50 o mediana): El 50% de los pedidos tienen un precio total de hasta R$86.90. Esto significa que la mitad de los pedidos tienen precios por debajo de este valor.

- 75% (percentil 75): El 75% de los pedidos tienen un precio total de hasta R$149.90. Esto indica el punto donde comienza el cuarto superior de los precios más altos.

- Max (máximo): El precio máximo de un pedido es de R$13,440. Esto podría ser un pedido excepcionalmente grande o que incluye productos de alto valor.

```python
# Visualizar la distribución de los precios de pedidos mediante un histograma
plt.hist(order_value['total_price_by_order'], bins=50, range=(0, 1000), color='skyblue', edgecolor='black')

# Lineas de mean y median
plt.axvline(order_value['total_price_by_order'].mean(), color='blue', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(order_value['total_price_by_order'].median(), color='green', linestyle='dashed', linewidth=1, label='Median')

# Etiquetas
plt.legend()
plt.xlabel('Precio')
plt.ylabel('Frecuencia')  # Número de pedidos en cada bin del histograma
plt.title('Valor de los Pedidos - Histograma', fontsize=12, fontweight='bold')
```

Podemos observar que el número de observaciones o pedidos tiende a disminuir a medida que aumentan los valores de price. El histograma muestra una distribución asimétrica, sesgada hacia la derecha, con una concentración de observaciones en los valores más bajos y una disminución gradual a medida que los valores de price aumentan. En una distribución sesgada a la derecha como esta, la media tiende a ser mayor que la mediana debido a la influencia de los valores extremos en la cola derecha, tal como podemos observar en el histograma. Esto sucede porque la media es muy sensible a la presencia de valores extremos, mientras que la mediana, por otro lado, es la posición central de los datos ordenados y no se ve afectada por los valores extremos en la misma medida que la media.

Esta forma de distribución sesgada hacia la derecha es común en datos de precios, ya que es más probable encontrar precios más bajos que precios muy altos en general. En el contexto de pedidos de un ecommerce, es común tener una amplia variedad de productos con diferentes precios. Esto puede resultar en una mayor cantidad de productos con precios bajos o asequibles y una menor cantidad de productos con precios muy altos considerados como más exclusivos o lujosos.

Esta distribución en los precios de los pedidos también puede deberse al comportamiento de compra de los clientes, donde los clientes tienden a comprar más productos de precios bajos que productos de precios altos debido a que, en mercados masivos, se venden productos de consumo diario o productos populares a precios asequibles.

#### ¿Cuál es el valor promedio del pedido (AOV)?

El Valor Promedio del Pedido (AOV, Average Order Value) es una métrica utilizada en análisis de ventas y comercio electrónico para medir cuánto, en promedio, gasta un cliente en cada compra realizada en una plataforma o tienda en línea. 

El AOV responde a la pregunta: "¿Cuánto gasta en promedio un cliente en cada pedido que realiza?" Un AOV más alto generalmente indica que los clientes están gastando más en cada compra, lo que puede ser un indicador de clientes satisfechos o de una estrategia de ventas efectiva.

El AOV se calcula sumando los precios de todos los productos en cada pedido y luego calculando el promedio de esos totales de pedido. Este valor representa el monto promedio que un cliente gasta en cada compra realizada en OLIST.

```python
# Calcular valor promedio del pedido (AOV) de las vents totales

print(F'El valor medio del pedido en las ventas totales (AOV) es R$ {items.groupby('order_id')['price'].sum().mean().round(2)}')
```

#### ¿Cuál es la tasa de crecimiento de pedidos mes a mes?


Debido a las inconsistencias identificadas en el año 2016 y en los ultimos meses del año 2018 hemos decidido imitar el periodo de tiempo de Enero de 2017 a Agosto de 2018 para un mejor análisis y una comparación más confiable y significativa entre los meses.

```python
# Limitar el periodo de tiempo de Enero de 2017 a Agosto de 2018
temporal_analysis_data = orders[(orders['order_purchase_year'] >= 2017) & 
                                        (orders['order_purchase_year'] <= 2018) & 
                                        ((orders['order_purchase_year'] != 2018) | 
                                         (orders['order_purchase_month'] <= 8))]

display(temporal_analysis_data.head(5))
```

```python
# Tasa de crecimiento de pedidos mes a mes (periodo Enero 2017 a Agosto 2018)

# Agrupar por año y mes de compra y contar el número de pedidos para cada mes
orders_by_year_and_month = temporal_analysis_data.groupby(['order_purchase_year', 'order_purchase_month', 'month_year'])['order_id'] \
    .nunique().reset_index().rename(columns= {'order_id':'n_orders'})

# Crear una copia de 'orders_by_year_and_month' en un nuevo DataFrame
monthly_order_growth_rate = orders_by_year_and_month.copy()

# Crear columa 'previous_month_orders' para poder realizar el calculo de la tasa de crecimiento
monthly_order_growth_rate['previous_month_orders'] = monthly_order_growth_rate['n_orders'].shift(1)

# Calcular tasa de crecimiento y agregar en nueva columna
monthly_order_growth_rate['growth_rate_%'] = ((monthly_order_growth_rate['n_orders'] - monthly_order_growth_rate['previous_month_orders']) \
                                           / monthly_order_growth_rate['previous_month_orders'] * 100).round(2).fillna(0)

print('Tasa de crecimiento de pedidos mes a mes:')
display(monthly_order_growth_rate.iloc[:, [0, 2, 5]])
```

#### ¿Cuál es la tasa de crecimiento de pedidos anual?


Para calcular una tasa de crecimiento anual precisa y significativa, se deben comparar períodos equivalentes en términos de duración y estacionalidad.

En este caso, comparar el mismo período de tiempo en ambos años, es decir, de enero a agosto de 2017 con enero a agosto de 2018. Esta comparación es más adecuada porque mantiene las condiciones estacionales y de mercado similares, lo que permitirá obtener una tasa de crecimiento más representativa y justa.

```python
# Limitar el periodo de tiempo a Enero-Agosto de 2017 y Enero-Agosto de 2018 para un mejor análisis y comparación

temporal_analysis_data_jan_to_aug = temporal_analysis_data[(temporal_analysis_data['order_purchase_year'] >= 2017) & 
                                        (temporal_analysis_data['order_purchase_year'] <= 2018) & 
                                         (temporal_analysis_data['order_purchase_month'] <= 8)]

display(temporal_analysis_data_jan_to_aug.head(5))
```

```python
# Calcular el número de pedidos por año en periodos equivalentes de 2017 y 2018
orders_by_year = temporal_analysis_data_jan_to_aug.groupby('order_purchase_year')['order_id'].nunique().reset_index()\
    .rename(columns= {'order_id':'total_orders'})

# Calcular tasa de crecimiento anual de pedidos

# Valor inicial (número de pedidos en 2017)
initial_value = orders_by_year.loc[0, 'total_orders']
# Valor final (número de pedidos en 2018)
final_value = orders_by_year.loc[1, 'total_orders']

# Calcular tasa de crecimiento anual
annual_order_growth_rate = ((final_value - initial_value) / initial_value * 100).round(2)

print(F'Número de pedidos por año:')
display(orders_by_year)
print(F'La tasa de crecimiento anual de pedidos fue de {annual_order_growth_rate}%')
```

#### ¿Cuál es el porcentaje de satisfacción del cliente (en base al CSAT)?

El Índice de Satisfacción del Cliente (CSAT) es una métrica que se utiliza para medir el nivel de satisfacción de los clientes con respecto a un producto, servicio o experiencia en general. Se calcula generalmente mediante encuestas en las que se pregunta a los clientes qué tan satisfechos están con diferentes aspectos de su experiencia.

En Olist Store, una vez que el cliente recibe el producto, o vence la fecha estimada de entrega, el cliente recibe por correo electrónico una encuesta de satisfacción donde pueden calificar su experiencia de compra con una escala de 1 a 5 y anotar algunos comentarios.

EL CSAT representa el porcentaje de clientes que están satisfechos con el producto o servicio, es decir, se calcula como el porcentaje de valoraciones positivas sobre el total de valoraciones. Este valor nos dará el porcentaje de clientes que están satisfechos con su experiencia de compra en Olist Store.




```python
# Determinar la escala de satisfacción del cliente de cada puntuación de reseña
reviews['review_scale'] = np.select([reviews['review_score'] == 5,
                                     reviews['review_score'] == 4,
                                     reviews['review_score'] == 3,
                                     reviews['review_score'] == 2,
                                     reviews['review_score'] == 1],
                                    ['positive', 
                                     'positive', 
                                     'neutral', 
                                     'negative', 
                                     'negative'])     

# Calcular el indice de satisfacción del cliente (CSAT)

# Número de valoraciones positivas
positive_reviews = reviews['review_scale'][reviews['review_scale'] == 'positive'].count()
# Total de valoraciones 
total_reviews = reviews['review_score'].count()

# Calcular CSAT
customer_satisfaction_score = (positive_reviews / total_reviews * 100).round(2)

print(F'El porcentaje satisfacción del cliente en general es {customer_satisfaction_score}%')                               
```
